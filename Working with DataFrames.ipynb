```python
import pandas as pd
```

```python
#extract data from csv
df = pd.read_csv("disney-characters.csv", 
                 sep=",", 
                 error_bad_lines=False, 
                 index_col=0)

print(df.head())
```

```python
#showing first 10 rows
df.head(10)

#showing last 10 rows
df.tail(10)
```

```python
#Accessing a column as a series:
df['movie_title']
```

```python
#Accessing a column as a data frame:
df[['movie_title']]
```

```python
#Accessing several rows:
df[['movie_title', 'hero', 'villain']]
```

```python
#Accessing a row (two ways)
df.iloc[1]
```

```python
#Accessing a row (two ways)
df.loc[1] #by index label
```

```python
#Accessing several rows
df.iloc[[0,5,10],:] #here, ":" means all columns
```

```python
#Accessing a range of rows
df.loc[3:7,:]
```

```python
#Accessing a single cell:
df.iloc[1,3]
```

```python
#Or:
df.loc[11,'villain']
```

```python
#Accessing a range of rows and columns (i.e. slicing the table):
df.iloc[13:18,0:4]
```

```python
#Or:
df.loc[13:17,"movie_title":"villain"]
```
